```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/biodatascience/datasci611/gh-pages/data/project2_2019/DISABILITY_ENTRY_191102.tsv", sep='\t',
                       usecols=['Client ID', 'Disability Type (Entry)', 'Disability Determination (Entry)'],
                       na_values=['**'], header = 0)
df[(df['Client ID'] == 190642) & (df['Disability Type (Entry)'] == 'Chronic Health Condition (HUD)')]
```

```python
df2 = df[df['Disability Determination (Entry)'] == 'Yes (HUD)'].drop_duplicates()
df2
```

```python
df3 = pd.pivot(df2, index = 'Client ID', columns = 'Disability Type (Entry)', values = 'Disability Determination (Entry)').fillna("No (HUD)")
df3
```
